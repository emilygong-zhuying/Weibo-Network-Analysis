### Building a network with self-edges (not recommended)

```python
# Subset the DataFrame to only include username and retweet_id
user_retweet = retweets_df_removed.loc[:, ("id", "username", "retweet_id")].copy()
user_retweet['retweet_username'] = user_retweet['retweet_id'].map(retweet_username_dict)
# user_retweet = user_retweet[~user_retweet['retweet_id'].isnull()]

# Building a self-edge for the posts being reposted
mask = user_retweet['retweet_id'].isnull()
user_retweet.loc[mask, 'retweet_username'] = user_retweet["username"]
# Build your graph
G=nx.from_pandas_edgelist(user_retweet, "username", "retweet_username")
```

```python
print("Graph info:", nx.info(G))
print("Graph density:", nx.density(G))
print("Graph is connected:", nx.is_connected(G))
print("Number of connected components:", nx.number_connected_components(G))
```

### Now, without self-edges

```python
# Subset the DataFrame to only include username and retweet_id
user_retweet2 = retweets_df_removed.loc[:, ("id", "username", "retweet_id")].copy()
user_retweet2['retweet_username'] = user_retweet2['retweet_id'].map(retweet_username_dict)
# user_retweet2 = user_retweet2[~user_retweet2['retweet_id'].isnull()]
# Build your graph
G2=nx.from_pandas_edgelist(user_retweet2, "username", "retweet_username")

print("Graph info:", nx.info(G2))
print("Graph density:", nx.density(G2))
print("Graph is connected:", nx.is_connected(G2))
print("Number of connected components:", nx.number_connected_components(G2))
```

### Why are the graphs different?

```python
print("The dataframe has length: ", len(retweets_df_removed))
print("Number of unique post id's: ", len(pd.unique(user_retweet2.loc[:, ('id')])))
print("Number of unique usernames: ", len(pd.unique(user_retweet2.loc[:, ('username')])))
print("Number of unique retweet id's: ",len(pd.unique(user_retweet2.loc[:, ('retweet_id')]))) # this includes nan as a unique value
print("Number of unique retweet usernames: ",len(pd.unique(user_retweet2.loc[:, ('retweet_username')]))) 
```

The stats above explains why we would have different simple networks when changing Id's to Username's. All the id's are unique (same length as the entire dataset), but not all the usernames. This makes sense because 1 user may repost many different posts. It would also make more logical sense to track users instead of posts, since we want to see the interaction between users behind each post rather than just an individual post.

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
    fig.subplots_adjust(hspace=0.05)  # adjust space between axes
    
    # plot the same data on both axes
    ax1.hist(degrees, bins = 100)
    ax2.hist(degrees, bins = 100)
    
    # zoom-in / limit the view to different portions of the data
    ax1.set_ylim(70, 20000)  # outliers only
    ax2.set_ylim(0, 70)  # most of the data

    ax1.xaxis.tick_top()
    ax1.tick_params(labeltop=False)  # don't put tick labels at the top
    ax2.xaxis.tick_bottom()

    d = .5  # proportion of vertical to horizontal extent of the slanted line
    kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
                  linestyle="none", color='k', mec='k', mew=1, clip_on=False)
    ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
    ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)
    
    plt.xlabel("Degree")
    plt.ylabel("Count")
    plt.title("Count vs Degree (title need to be fixed)")
    plt.show()

plot_degree_dist(G2)
```

```python
def overlap(centrality_list1, centrality_list2):
    list1 = []
    list2 = []
    overlap = []
    for i in centrality_list1:
        list1.append(i[0])
    for j in centrality_list2:
        list2.append(j[0])
    for k in list1:
        if k in list2:
            overlap.append(k)
    return overlap
        
print("There is a ", len(overlap(sorted_betweenness[1:21], sorted_eigenvector[1:21]))/20*100, "% overlap between betweenness centrality and eigenvector centrality for the top 20 users, namely: ", overlap(sorted_betweenness[1:21], sorted_eigenvector[1:21]))
print("There is a ", len(overlap(sorted_betweenness[1:21], sorted_random_walk[1:21]))/20*100, "% overlap between betweenness centrality and random walk centrality for the top 20 users, namely: ", overlap(sorted_betweenness[1:21], sorted_random_walk[1:21]))
print("There is a ", len(overlap(sorted_eigenvector[1:21], sorted_random_walk[1:21]))/20*100, "% overlap between eigenvectror centrality and random walk centrality for the top 20 users, namely: ", overlap(sorted_eigenvector[1:21], sorted_random_walk[1:21]))
```

```python
import demoji
def clean(context):
    if len(context) == 0:
        return ""
    # first get rid of emojis
    context = demoji.replace(context, repl="")
    
    # not getting rid of the username after @
    at = re.findall("@", context)
    context = re.sub("@",",",context)

    # tags = re.findall("#(.{0,30})#",context)
    # context = re.sub("#.{0,30}#","",context)

    tags = re.findall("#",context)
    context = re.sub("#",",",context)

    # title = re.findall("【(.{0,30})】",context)
    # context = re.sub("【.{0,30}】","",context)

    title = re.findall("【",context)
    context = re.sub("【",",",context)

    title = re.findall("】",context)
    context = re.sub("】",",",context)

    parentheses = re.findall("（(.{0,8})）",context)
    context = re.sub("（.{0,8}）","",context)

    parentheses = re.findall("（(.{0,30})）",context)
    context = re.sub("（.{0,30}）","",context)
    
    ch = 'L'
    pattern  = ch + ".*"
    context = re.sub(pattern, '', context)

    context = re.sub("（", ",", context)
    
    '''
    context = re.sub(r'[^\w\s]',',',context) 
    this line helps clean chinese punctuations by replacing everything to ,
    not the best -> will find another way to clean it later
    '''
    if len(context) == 0:
        return ""

    if context[0] == ',':
        context = context[1:]
        
    r = re.compile(r'([.,/#!$%^&*;:{}=_`~()-])[.,/#!$%^&*;:{}=_`~()-]+')
    context = r.sub(r'\1', context)
    
    if len(context) == 0:
        return ""
    
    if context[0] == ',':
        context = context[1:]
    
    return context
```

# Manually checking sentiment

Out of 3971 original posts, I randomly selected 10 samples of 100 posts (with replacement). Note we use random_state = 1 to ensure the reproducibility of the examples.

Since the sentiment analysis only concerns the post_text and username(for identification), we will only keep those 2 cols.

2 for positive, 1 for neutral, 0 for negative

My personal bias is unavoidable in this process.

```python
selected_rows['sentiment_score'] = 1

def samples(df, n):
    final_sample = pd.DataFrame()
    for i in range(n):
        temp = df[['username', 'post_text','sentiment_score']].sample(n = 100, replace = True, random_state = i).copy()
        final_sample = final_sample.append(temp)   
    return final_sample

samples_df = samples(selected_rows, 2)
samples_df.to_csv("manual_sa.csv")
```
